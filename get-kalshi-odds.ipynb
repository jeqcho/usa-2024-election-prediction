```python
# List of state abbreviations
state_abbr = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 
    'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 
    'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 
    'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 
    'WI', 'WY'
]

# Base URL without the state abbreviation
base_url = "https://kalshi.com/markets/presparty"

# Generate the full URLs
state_urls = [f"{base_url}{abbr.lower()}" for abbr in state_abbr]
```

```python
# Step 1: Install Selenium and dependencies
!apt-get update
!apt install -y chromium-chromedriver
!pip install selenium

# Step 2: Configure ChromeDriver for Colab
import sys
sys.path.insert(0, '/usr/lib/chromium-browser/chromedriver')

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# Step 3: Set up Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ensure Chrome runs in headless mode
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-gpu")
```

```python
# Provide the path to your ChromeDriver 
driver = webdriver.Chrome(options=chrome_options)

# Function to scrape data from a specific state's market page
def scrape_state_data(state_url):
    print(state_url)
    driver.get(state_url)
    time.sleep(3)  # Wait for the page to fully load
    
    try:
        # Extract the elements using the text tags (as you've mentioned)
        elements = driver.find_elements(By.TAG_NAME, 'text')

        # Extract the required data using indices (6 for leading candidate, 7 for odds, etc.)
        if len(elements) > 9:
            leading_candidate = elements[6].text.strip()
            leading_candidate_odds = elements[7].text.strip()
            losing_candidate = elements[8].text.strip()
            losing_candidate_odds = elements[9].text.strip()

            print(f"Leading Candidate: {leading_candidate}, Odds: {leading_candidate_odds}")
            print(f"Losing Candidate: {losing_candidate}, Odds: {losing_candidate_odds}")
            
            return {
                "state_url": state_url,
                "leading_candidate": leading_candidate,
                "leading_odds": leading_candidate_odds,
                "losing_candidate": losing_candidate,
                "losing_odds": losing_candidate_odds
            }
        else:
            print(f"Insufficient data found for {state_url}")
            return None
    except Exception as e:
        print(f"Error scraping {state_url}: {str(e)}")
        return None

# List of state URLs (you can use the earlier script to generate this list for all states)
# Scrape odds for all states
all_state_data = []
for url in state_urls:
    state_data = scrape_state_data(url)
    if state_data:
        all_state_data.append(state_data)

# Close the browser
driver.quit()

# Print all the results
for data in all_state_data:
    print(data)
```

```python
import pandas as pd

# Assuming 'all_state_data' has been collected already

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(all_state_data)

# Filter out rows where the leading candidate is "Harris" or "Trump"
filtered_df = df[~df['leading_candidate'].str.contains('Harris|Trump', case=False, na=False)]

# Display the filtered DataFrame
filtered_df
```

```python
# Assuming we have the old DataFrame (df) which needs to be updated with the new data

# Define the updated states DataFrame again
new_data = {
    "state_url": [
        'https://kalshi.com/markets/prespartyaz', 
        'https://kalshi.com/markets/prespartyga', 
        'https://kalshi.com/markets/prespartyin', 
        'https://kalshi.com/markets/prespartyms', 
        'https://kalshi.com/markets/prespartync', 
        'https://kalshi.com/markets/prespartywy'
    ],
    "leading_candidate": ["Donald Trump", "Donald Trump", "Donald Trump", "Donald Trump", "Donald Trump", "Donald Trump"],
    "leading_odds": [72, 70, 99, 99, 70, 99],
    "losing_candidate": ["Kamala Harris", "Kamala Harris", "Kamala Harris", "Kamala Harris", "Kamala Harris", "Kamala Harris"],
    "losing_odds": [28, 30, 1, 1, 30, 1]
}

# Convert the new data into a DataFrame
new_df = pd.DataFrame(new_data)

# Assuming 'df' is the old DataFrame, let's replace the states matching the URLs in the new DataFrame
# First, drop the old entries matching the state URLs
df = df[~df['state_url'].isin(new_df['state_url'])]

# Append the new data to the old DataFrame
df = pd.concat([df, new_df], ignore_index=True)
```

```python
# Function to add "%" if not already present
def add_percentage(odds):
    if isinstance(odds, str) and '%' not in odds:
        return f"{odds}%"
    elif isinstance(odds, int) or isinstance(odds, float):
        return f"{odds}%"
    return odds

# Apply the function to the leading and losing odds columns
df['leading_odds'] = df['leading_odds'].apply(add_percentage)
df['losing_odds'] = df['losing_odds'].apply(add_percentage)

df
```

```python
# Extract the state name from the state_url and add it as a new column
df['state'] = df['state_url'].apply(lambda url: url.split('/presparty')[-1].upper())

df
```

```python
# Create a new column 'harris_odds' which copies the 'losing_odds' where 'losing_candidate' is 'Harris'
df['harris_odds'] = df.apply(lambda row: row['losing_odds'] if 'Harris' in row['losing_candidate'] else row['leading_odds'], axis=1)

df
```
